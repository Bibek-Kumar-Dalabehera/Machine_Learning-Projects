```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['target']=iris.target
```

```python
df[df.target==0].head()#Datasets of setosa
df[df.target==1].head()#Datasets of versicolor
df[df.target==2].head()#Datasets of virginica

#Create a flowername column as compare with target columns
df['flowers_name']=df.target.apply(lambda x:iris.target_names[x])
```

```python
df.head()
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
df.drop(columns=['target'],axis=1,inplace=True)
```

#### Graphical Representation

```python
#Sepal Length/Width
plt.xlabel('Sepal_length(cm)')
plt.ylabel('Sepal_width(cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],marker='D',color='black')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],marker='D',color='yellow')
plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],marker='D',color='blue')
plt.title('Sepal_length/width',fontweight='bold',color='red')
plt.show()
```

```python
#Petal Length/Width
plt.xlabel('Petal_length(cm)')
plt.ylabel('Petal_width(cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],marker='D',color='black')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],marker='D',color='yellow')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],marker='D',color='blue')
plt.title('Petal_length/width',fontweight='bold',color='blue')
plt.show()
```

##### Kdeplot

```python
sn.kdeplot(data=df,fill=True,hatch="/")
plt.title("Density vs Flowers Name",fontweight='bold',color='magenta')
plt.show()
```

##### Box Plot

```python
sn.boxplot(x=df.flowers_name,y=df['sepal width (cm)'])
plt.title("Flowers Name with Sepal width(cm)",fontweight='bold',color='red')
plt.show()
sn.boxplot(x=df.flowers_name,y=df['sepal length (cm)'])
plt.title("Flowers Name with Sepal length(cm)",fontweight='bold',color='green')
plt.show()
sn.boxplot(x=df.flowers_name,y=df['petal width (cm)'])
plt.title("Flowers Name with Petal width(cm)",fontweight='bold',color='blue')
plt.show()
sn.boxplot(x=df.flowers_name,y=df['petal length (cm)'])
plt.title("Flowers Name with Petal length(cm)",fontweight='bold',color='orange')
plt.show()
```

##### Count Plot
