```python
import pandas as pd
import numpy as np
```

##### Loading Data set and creation of dataframe 

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['Target']=iris.target
```

```python
iris.target_names
```

```python
df.Target.unique()
```

```python
df[df.Target==0].head()#Datasets of setosa
df[df.Target==1].head()#Datasets of versicolor
df[df.Target==2].head()#Datasets of virginica
```

```python
#Create a flowername column as compare with target columns
df['flowers_name']=df.Target.apply(lambda x:iris.target_names[x])
df.head()
```

```python
df.drop(['Target'],axis=1,inplace=True)
```

```python
df.head()
```

##### Label Encoding

```python
x=df.drop(['flowers_name'],axis=1)
y=df.flowers_name
```

```python
x.head(2)
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
y=le.fit_transform(y)
```

```python
y
```

##### ML Implementation

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
from sklearn.metrics import accuracy_score
models={
    "LogisticzRegression":LogisticRegression(),
    "DecissionTreeRegressor":DecisionTreeClassifier(),
    "RandomForestRegressor":RandomForestClassifier(),
    "SVR":SVC()
}
for i in range(len(list(models))):
    model=list(models.values())[i]
    print(model)
    model.fit(x_train,y_train)

    #Make prediction
    y_train_pred=model.predict(x_train)
    y_test_pred=model.predict(x_test)

    #Training set score
    model_train_as=accuracy_score(y_train,y_train_pred)


    #Testomg set score
    model_test_as=accuracy_score(y_test,y_test_pred)
    
    
    print("Model Performance Of Traing Set")
    print('Accuracy:',model_train_as)
    

    print('\n')

    print("Model Performance Of Testing Set")
    print('Accuracy:',model_test_as)

    print('\n--------------------------------------\n')
```

Here we can see except LogisticRegression model all model gives 100% accuracy so we can select any of these
