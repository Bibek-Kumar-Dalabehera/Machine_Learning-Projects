Here we claerly see that some data in total sqrft in range format so I decided to covert into single and I do like this <br>

```python
def converrage(x):
    temp=x.split('-')
    if len(temp)==2:
        return (float(temp[0]) + float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
data.total_sqft=data['total_sqft'].apply(converrage)
```

```python
def  remove_outlier_sqrft(df):
    df_output=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sqrft)

        st=np.std(subdf.price_per_sqrft)

        gen_df=subdf[(subdf.price_per_sqrft > (m-st)) & (subdf.price_per_sqrft <= (m+st))]
        df_output=pd.concat([df_output,gen_df],ignore_index=True)
    return df_output

data = remove_outlier_sqrft(data)
data.describe()
```

```python
def bhk_outlier_remover(df):
    exclude_indecies=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_stats={}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats={
                'mean':np.mean(bhk_df.price_per_sqrft),
                'std':np.std(bhk_df.price_per_sqrft),
                'count':bhk_df.shape[0]
            }

        for bhk,bhk_df in location_df.groupby('bhk'):
            stats=bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indecies=np.append(exclude_indecies,bhk_df[bhk_df.price_per_sqrft<(stats['mean'])].index.values)
    return df.drop(exclude_indecies,axis='index')
```

```python
data=bhk_outlier_remover(data)
```

#### ML implementation

```python
X=data.drop(columns=['price'])
y=data.price
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score,accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

##### Linear Regression

```python
column_trnas=make_column_transformer((OneHotEncoder(sparse_output=False),['location']),remainder='passthrough')
```

```python
scclar = StandardScaler()
```

```python
lr=LinearRegression(fit_intercept=True)
```

```python
pipe1=make_pipeline(column_trnas,scclar,lr)
```

```python
pipe1.fit(X_train,y_train)
```

```python
y_pred_lr= pipe1.predict(X_test)
r2_score(y_test,y_pred_lr)
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor()
pipe=make_pipeline(column_trnas,scclar,model)
pipe.fit(X_train,y_train)
pipe.score(X_test,y_test)
```
